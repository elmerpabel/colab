```python
# 2. Importar librerías necesarias
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
# 3. Cargar el archivo CSV que subiste
df = pd.read_csv('/content/Escuelaglobal/train_adultos.csv',sep=',')
```

```python
# 4. Generar el perfil
profile = ProfileReport(df, title="Reporte de Profiling - train_adultos", explorative=True)
```

```python
# 5. Mostrar el reporte dentro de Colab
profile.to_notebook_iframe()
```

```python
# 6. (Opcional) Exportar el reporte a HTML
profile.to_file("reporte_train_adultos.html")
```
